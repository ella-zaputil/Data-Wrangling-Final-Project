## Data Source 1

Downloaded "IMDB Movies Dataset" CSV file from Kaggle 

```python
#Import needed libraries
import pandas as pd

#Read in CSV file as data frame
imdb_df = pd.read_csv("imdb_top_1000.csv", encoding = 'utf-8')

#Display IMDB data frame
display(imdb_df)
```

```python
#Drop Poster_Link, Certificate, Metascore, Overview, No_Of_Votes columns. They are not valuable to my analysis.
#Gross revenue will be collected from web scraping later
imdb_df = imdb_df.drop(columns = ["Poster_Link", "Certificate", "Meta_score", "Overview", "No_of_Votes", "Gross"])
display(imdb_df)
```

```python
#Rename columns to make more sense for overall analysis
#Series_Title -> Movie Title
#Released_Year -> Release Year

imdb_df = imdb_df.rename(columns = {'Series_Title':'Movie Title', 'Released_Year':'Release Year'})
display(imdb_df)
```

```python
#Analyze data types
print(imdb_df.dtypes)
```

```python
#Clean data, convert needed data types 

#Release year: object -> integer
imdb_df['Release Year'] = pd.to_numeric(imdb_df['Release Year'], errors = 'coerce').astype('Int64')

#Runtime: object -> integer
imdb_df['Runtime'] = imdb_df['Runtime'].str.replace(' min', '').astype('Int64')

#Recheck data types
print(imdb_df.dtypes)
```

## Data Source 2

Scraped data from https://www.the-numbers.com/movie/budgets/all  
Scraped data for 1-100, 600-800 ranked movies

```python
#Import needed libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager 

# needed to randomize time on the page and scrolling
import time
import random
```

```python
# initialize the Selenium web driver (Chrome in this case)
browser = webdriver.Chrome()

#Create lists to store data
rankings = []
release_dates = []
movies = []
production_budgets = []
domestic_gross = []
worldwide_gross = []

base_url = 'https://www.the-numbers.com/movie/budgets/all'
browser.get(base_url)  # Open the page in the browser

wait_time = random.uniform(3, 10)  # random wait time between 3 and 10 seconds
time.sleep(wait_time)

# Find the table containing the data
table = browser.find_element(By.XPATH, "//table")

# Get the rows of the table (skip the first row, which contains the headers)
for tr in table.find_elements(By.TAG_NAME, 'tr')[1:]:
    cells = tr.find_elements(By.TAG_NAME, 'td')

    # Extract the relevant data from each row and append to the lists
    if len(cells) > 1:  # Make sure it's not an empty row
        ranking = cells[0].text.strip()
        release_date = cells[1].text.strip()
        movie = cells[2].text.strip()
        production_budget = cells[3].text.strip().replace(',', '').replace('$','')  # Remove commas and $
        domestic = cells[4].text.strip().replace(',', '').replace('$','')  # Remove commas and $
        worldwide = cells[5].text.strip().replace(',', '').replace('$','')  # Remove commas and $

        # Append the cleaned data to the corresponding lists
        rankings.append(ranking)
        release_dates.append(release_date)
        movies.append(movie)            
        production_budgets.append(production_budget)
        domestic_gross.append(domestic)
        worldwide_gross.append(worldwide)
    
# Close the browser after scraping
browser.quit()

# Convert the lists into a pandas DataFrame
numbers_data = {
    'Ranking': rankings,
    'Release Date': release_dates,
    'Movie': movies,
    'Production Budget': production_budgets,
    'Domestic Gross': domestic_gross,
    'Worldwide Gross': worldwide_gross
}

numbers_df = pd.DataFrame(numbers_data)
display(numbers_df)
```

```python
browser = webdriver.Chrome()

#Create lists to store data
rankings2 = []
release_dates2 = []
movies2 = []
production_budgets2 = []
domestic_gross2 = []
worldwide_gross2 = []

base_url = 'https://www.the-numbers.com/movie/budgets/all'
page_number = 101

while page_number < 1901: 
    next_url = f"{base_url}/{page_number}"  # Construct the URL for each page
    browser.get(next_url)  # Open the page in the browser
    
    wait_time = random.uniform(3, 10)  # random wait time between 3 and 10 seconds
    time.sleep(wait_time)

    # Find the table containing the data
    table = browser.find_element(By.XPATH, "//table")

    # Get the rows of the table (skip the first row, which contains the headers)
    for tr in table.find_elements(By.TAG_NAME, 'tr')[1:]:
        cells = tr.find_elements(By.TAG_NAME, 'td')

        # Extract the relevant data from each row and append to the lists
        if len(cells) > 1:  # Make sure it's not an empty row
            ranking = cells[0].text.strip()
            release_date = cells[1].text.strip()
            movie = cells[2].text.strip()
            production_budget = cells[3].text.strip().replace(',', '').replace('$','')  # Remove commas and $
            domestic = cells[4].text.strip().replace(',', '').replace('$','')  # Remove commas and $
            worldwide = cells[5].text.strip().replace(',', '').replace('$','')  # Remove commas and $

            # Append the cleaned data to the corresponding lists
            rankings2.append(ranking)
            release_dates2.append(release_date)
            movies2.append(movie)            
            production_budgets2.append(production_budget)
            domestic_gross2.append(domestic)
            worldwide_gross2.append(worldwide)

    page_number += 100

# Close the browser after scraping
browser.quit()

# Convert the lists into a pandas DataFrame
numbers_df2 = pd.DataFrame({
    'Ranking': rankings2,
    'Release Date': release_dates2,
    'Movie': movies2,
    'Production Budget': production_budgets2,
    'Domestic Gross': domestic_gross2,
    'Worldwide Gross': worldwide_gross2
})

numbers_df = pd.concat([numbers_df, numbers_df2], ignore_index = True)
display(numbers_df)
```

```python
#Check data types
print(numbers_df.dtypes)
```

```python
#Change needed data types

#Ranking: object -> integer
numbers_df['Ranking'] = pd.to_numeric(numbers_df['Ranking'], errors = 'coerce').astype('Int64')

#Release Date: object -> date
numbers_df['Release Date'] = pd.to_datetime(numbers_df['Release Date'], format='%b %d, %Y')

#Production Budget: object -> integer
numbers_df['Production Budget'] = pd.to_numeric(numbers_df['Production Budget'], errors = 'coerce').astype('Int64')

#Domestic Gross: object -> integer
numbers_df['Domestic Gross'] = pd.to_numeric(numbers_df['Domestic Gross'], errors = 'coerce').astype('Int64')

#Production Budget: object -> integer
numbers_df['Worldwide Gross'] = pd.to_numeric(numbers_df['Worldwide Gross'], errors = 'coerce').astype('Int64')

#Recheck data types
print(numbers_df.dtypes)
```

```python
#Extract the year from release date to use to join
numbers_df['Release Year'] = numbers_df['Release Date'].dt.year
```

```python
#Rename Movie column to match IMDB data frame
numbers_df.rename(columns={'Movie': 'Movie Title'}, inplace=True)

display(numbers_df)
```
